```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)
```

```python
n1.next = n2
n2.next = n3
n3.next = n4
```

```python
head = n1
```

```python
curr = head
total = 0
while curr is not None:
    total += curr.data
    curr = curr.next
print("Sum =", total)
```

```python
curr = head
maximum = head.data
while curr is not None:
    if curr.data > maximum:
        maximum = curr.data
    curr = curr.next
print("Maximum =", maximum)
```

```python
prev = None
curr = head
```

```python
while curr is not None:
    temp = curr.next
    curr.next = prev
    prev = curr
    curr = temp
```

```python
head = prev
```

```python
print("Reversed List:")
curr = head
while curr is not None:
    print(curr.data, end=" -> ")
    curr = curr.next
print("None")
```

```python
class Student:
    def __init__(self, id, name, result):
        self.id = id
        self.name = name
        self.result = result

    def std_info(self):
        print("ID:", self.id)
        print("Name:", self.name)
        print("Result:", self.result)
```

```python
n= int(input("enter how many students"))
students=[]
```

```python
for i in range(n):
    id = int(input("Enter student ID: "))
    name = input("Enter student name: ")
    result = input("Enter student result: ")
    s1 = Student(id, name, result)
    students.append(s1)
```

```python
for s1 in students:
    s1.std_info()
```
