```python
class Cart:
    def __init__(self, name, code, price, qty):
        self.name = name
        self.code = code
        self.price = price
        self.qty = qty
        self.total = self.price * self.qty

    def info(self):
        print("Name:", self.name)
        print("Code:", self.code)
        print("Price:", self.price)
        print("Quantity:", self.qty)
        print("Total:", self.total)
        print()


c1 = Cart("Laptop", 1122, 50000, 2)
c2 = Cart("Phone", 23234, 25000, 1)

c1.info()
c2.info()
```

```python
class Calculator:
    def __init__(self,a,b,):
        self.a=a
        self.b=b
    def add(self):
        return self.a+self.b
    def sub(self):
        return self.a-self.b
    def mul(self):
        return self.a*self.b
    def div(self):
        return self.a%self.b
```

```python
c=Calculator(10,30)
```
