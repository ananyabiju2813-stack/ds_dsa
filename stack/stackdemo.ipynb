```python
class Stack:
    def __init__(self):
        self.data=[]
    def push(self,item):
        self.data.append(item)
    def display(self):
        for i in range(-1,-(len(self.data)+1), -1):
         print(self.data[i],sep=" ")
    def pop(self):
        if self.is_empty()!=True:
            return self.data.pop()
        else:
            print("stack is empty")
    def is_empty(self):
        if len(self.data)==0:
            return True
```

```python
s=Stack()
s.push(10)
s.push(20)
s.push(30)
s.push(40)
s.push(50)
s.pop()
s.pop()
s.pop()
s.pop()
s.pop()
```

```python
print(s.data)
```

```python
s.display()
```

```python
s.pop()
```

```python
class Order:
    def __init__(self,id,date,price,qty):
     self.id=id
     self.date=date
     self.price=price
     self.qty=qty
    def order_info(self):
        print(f"orderid:{self.id}orderdate:{self.date}orderprice:{self.price}orderqty:{self.qty}")
```

```python
o1=Order(100,"11-2-2025",200,2)
o2=Order(200,"31-4-2025",300,9)
o3=Order(300,"21-5-2025",204,2)
o4=Order(400,"12-2-2025",600,1)
o5=Order(500,"13-6-2025",500,3)
```

```python
from collections import deque
```

```python
order_queue=deque()
```

```python
order_queue.append(o1)
order_queue.append(o2)
order_queue.append(o3)
order_queue.append(o4)
order_queue.append(o5)
```

```python
while order_queue:
    cur=order_queue.popleft()
    print("processing order")
    cur.order_info()
```

```python
class patient:
    def __init__(self, id,name,dept,type):
        self.id=id
        self.name=name
        self.dept=dept
        self.type=type
    def patient_info(self):
        print(f"patientid:{self.id}patientname:{self.name}patientdept:{self.dept}patienttype:{self.type}")
```

```python
p1=patient(1,"abhi","cardio","out")
p2=patient(2,"ananya","cardio","in")
p3=patient(3,"emil","ent","out")
p4=patient(4,"arun","gastro","in")
p5=patient(5,"anu","gyn","out")
```

```python
import heapq
```

```python
patients=[]
```

```python
heapq.heappush(patients,(3,p1))
heapq.heappush(patients,(4,p2))
heapq.heappush(patients,(2,p3))
heapq.heappush(patients,(1,p4))
heapq.heappush(patients,(5,p5))
```

```python
while patients:
    pr,p=heapq.heappop(patients)
    p.patient_info()
```
