```python
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)
```

```python
n1.next = n2
n2.next = n3
n3.next = n4
```

```python
head = n1
```

```python
curr = head
total = 0
while curr is not None:
    total += curr.data
    curr = curr.next
print("Sum =", total)
```

```python
curr = head
maximum = head.data
while curr is not None:
    if curr.data > maximum:
        maximum = curr.data
    curr = curr.next
print("Maximum =", maximum)
```

```python
prev = None
curr = head
```

```python
while curr is not None:
    temp = curr.next
    curr.next = prev
    prev = curr
    curr = temp
```

```python
head = prev
```

```python
print("Reversed List:")
curr = head
while curr is not None:
    print(curr.data, end=" -> ")
    curr = curr.next
print("None")
```
