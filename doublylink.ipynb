```python
n1 = Node(500)
n2 = Node(600)
n3 = Node(700)
n4 = Node(400)
```

```python
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3
```

```python
cur= n1
max1 = float('-inf')
max2 = float('-inf')
print(max1)

while cur!= None:
    if cur.data > max1:
        max2 = max1
        max1 = cur.data
    elif cur.data > max2:
        max2 = cur.data
    cur = cur.next

print("Second Maximum =", max2)
```
